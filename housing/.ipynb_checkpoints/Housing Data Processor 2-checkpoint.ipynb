# Housing Data Analysis 2

The goal of this class of data analysis is as follows:

HousingPreprocessor
1. Separate data into numerical and categorical data. 
2. Impute and  linearly normalize all float type data into the interval [0,1].

HousingExtractor
3. We introduce a hyperparameter for each zero heavy data with dtype float, call this <b>zeroRatios</b> and another integer hyperparameter <b>numSigFeat</b>. We separate the numerical data into those whose ratio of zeros is higher than <b>zeroRatios</b> and those <= <b>zeroRatios</b>.
4. From the features with zero ratio < <b>zeroRatios</b>, we select at most <b>numSigFeat</b> number and call their indices cor_inds (correlated indices to SalePrice). The rest of the numerical features are called idp_inds (independent to SalePrice).

# 1. Load Data

```python
# Load the data
# data is first downloweded into DATA_PATH from 
# http://archive.ics.uci.edu/ml/machine-learning-databases/00267/data_banknote_authentication.txt
import os

DATA_PATH = 'data'
FILE_NAME = 'train.csv'

def load_data(data_path=DATA_PATH, file_name=FILE_NAME, ratio=0.9):
    # load everything into data
    file_path = os.path.join(data_path, file_name)
    data = pd.read_csv(file_path)
    
    slice_ind = int(data.shape[0]*ratio)
    
    return data.iloc[0:slice_ind], data.iloc[slice_ind:]
```

# 2. Preliminary processing
HousingPreprocessor:
1. Imputes all missing value using a specified strategy
2. Normalize all the data to (0,1) via MaxMin scaling by default
3. Take log of the SalePrice

```python
class HousingPreprocessor(object):
    def __init__(self, 
                 imputer=SimpleImputer(), 
                 normalizer=MinMaxScaler()):
        self.imputer = imputer
        self.normalizer = normalizer
        self.num_inds = [] # does not include sale price
        self.cat_inds = []
    
    # separate_num_and_cat: self,
    # data does not include column ID
    def seperate_num_and_cat(self, data):
        # numerical data type
        cont_type = [int, np.int64, float, np.float64]

        for col in data.columns:
            if data[col].dtype in cont_type:
                self.num_inds.append(col)
                #data[col].to_numeric()
            else:
                self.cat_inds.append(col)
        
        self.num_inds.pop() # don't scale SalePrice

    
    # data includes all columns
    # data is not consumed, a copy is made
    def fit(self, data):
        X = data.copy()
        del X['Id']
        
        self.seperate_num_and_cat(X)
        
        X[self.num_inds] = self.imputer.fit_transform(X[self.num_inds])
        X[self.num_inds] = self.normalizer.fit(X[self.num_inds])
    
    # transform: self, pd.DataFrame --> pd.DataFrame
    # X includes all columns
    # X is consumed and modified
    # take log of SalePrice
    def transform(self, X):
        del X['Id']
        
        X['SalePrice'] = np.log(X['SalePrice'])
        
        X[self.num_inds] = self.imputer.transform(X[self.num_inds])
        X[self.num_inds] = self.normalizer.transform(X[self.num_inds])
        
        return X
    
    # fit_transform: self, pd.DataFrame --> pd.DataFrame
    # data includes all columns
    # data is consumed and modified
    # take log of SalePrice
    def fit_transform(self, X):
        del X['Id']
        
        X['SalePrice'] = np.log(X['SalePrice'])
        
        self.seperate_num_and_cat(X)
        
        X[self.num_inds] = self.imputer.fit_transform(X[self.num_inds])
        X[self.num_inds] = self.normalizer.fit_transform(X[self.num_inds])
        
        return X     
            
        
        
```

## 3.4 HousingExtractorQuad
input is a pd.DataFrame of numerical values X, output (fit_transform) is all possible quadratic terms in X.

```python
class HousingMakeQuad(object):
    def __init__(self):
        self.size = None
    
    # assumes that nums has no id and SalePrice column
    def fit(self, nums):
        self.size = nums.values.shape
    
    # transform: pd.DataFrame --> pd.DataFrame
    # nums has no id and SalePrice
    def transform(self, nums):
        X = nums.values
        a = X.reshape((self.size[0], self.size[1], 1))
        b = X.reshape((self.size[0], 1, self.size[1]))
        
        c = (a*b).reshape(self.size[0], self.size[1] ** 2)
        
        return pd.DataFrame(np.unique(c, axis=1))
    
    # fit_transform: pd.DataFrame --> pd.DataFrame
    def fit_transform(self, nums):
        self.fit(nums)
        return self.transform(nums)
```

# 3.3 HousingExtractor

```python
class HousingExtractor(object):
    def __init__(self, zeroRatio=0.9, numSigFeat=10, quad=False):
        self.zeroRatio = zeroRatio        
        self.numSigFeat = numSigFeat
        self.cov = None
        self.zeroRank = None
        self.quader = HousingMakeQuad()
        
        self.cor_inds = []
        self.idp_inds = []
        self.cat_inds = []       
        
        self.standardScaler = StandardScaler()
        
        self.quad = quad
    
    # assumes that data has no id column but has SalePrice columns
    # num_inds does not contain 'SalePrice'
    # data contains SalePrice
    def fit(self, data, num_inds, cat_inds, debug=False):
        X = data.copy()
        
        self.cat_inds = cat_inds
        
        if self.quad:
            nums = X[num_inds]
            nums = pd.concat([nums, quader.fit_transform(nums), pd.DataFrame(X['SalePrice'], columns=['SalePrice'])], axis=1)
            num_inds = nums.columns
        else:
            nums = X[num_inds + ['SalePrice']]
        
        
        self.zeroRank = ((nums > 0).sum()/nums.shape[0]).nlargest(1000)
        zeroLight_inds = self.zeroRank[self.zeroRank >= 1-self.zeroRatio].index
        zeroHeavy_inds = self.zeroRank[self.zeroRank < 1-self.zeroRatio].index
        
        nums = pd.DataFrame(self.standardScaler.fit_transform(nums), 
                                          columns=nums.columns)
        nums = nums[zeroLight_inds]
        if debug:
            print(zeroLight_inds)
        
        if debug:
            print(nums.describe())
        corrmat = np.abs(nums.cov())
        self.cor_inds = corrmat.nlargest(self.numSigFeat+1, 'SalePrice')['SalePrice'].index
        if debug:
            print(self.cor_inds)
            print(corrmat.nlargest(self.numSigFeat+1, 'SalePrice')['SalePrice'])
            
        self.cov = nums[self.cor_inds].cov()
        self.cor_inds = self.cor_inds.values[1:] # pop SalePrice

        for col in num_inds:
            if col not in self.cor_inds:
                self.idp_inds.append(col)
    
    # returned data does not contain Id and SalePrice
    def transform(self, X):
        return nums[self.cor_inds], nums[self.idp_inds], X[self.cat_inds]
    
    def fit_transform(self, data, num_inds, cat_inds):
        self.fit(data, num_inds, cat_inds)
        return self.transform(data)
    
    def show_cov(self):
        f, ax = plt.subplots(figsize=(12,10))
        sns.heatmap(np.abs(self.cov), vmax = 1, square=True);
```
