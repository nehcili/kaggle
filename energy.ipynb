```python
import pandas as pd
import numpy as np
data = pd.read_csv('energy.csv')
```

```python
energy_costs = data.iloc[:,645:671]
mean_energy_cost = energy_costs.mean()
```

```python
import matplotlib.pyplot as plt
f, ax = plt.subplots(figsize=(12,12))
x = plt.pie(mean_energy_cost.values, labels=mean_energy_cost.index)
```

```python
import seaborn as sns
```

```python
feature_list = data.columns[0:480]
feature_list = feature_list.append(pd.Index(['DOLELSPH']))
features = data[feature_list]
```

```python
corrmat = features.corr()

k = 20
cols = corrmat.nlargest(k, 'DOLELSPH')['DOLELSPH'].index
corrmat_small = features[cols].corr()
```

```python
f, ax = plt.subplots(figsize=(12,10))
sns.heatmap(corrmat_small, vmax = 0.8, square=True);
```
